Get the URL address of each memes

```python
import urllib.request
from bs4 import BeautifulSoup
import re
import time
#You can change the page3/4/5 etc of the url and then use the for loop to capture it
url = "https://knowyourmeme.com/memes/popular/page/"
file = open("knowyourmemes.txt","w")
linkMap = {}

page = 10 #Determine the range of Url pages to crawl by changing the number
for i in range(1,page+1):
    url1 = url + str(i)
    print(url1)
    html = urllib.request.urlopen(url1)
    bsObj = BeautifulSoup(html,'html.parser')

    for link in bsObj.find('div',id = "infinite-scroll-wrapper").find_all('a',href = re.compile("^(/memes/)((?!;|page)\S)*$")):
        href = link.attrs['href']
         # {
        #'/memes/slender-man'：1，
        # '/memes/navy-seal-copypasta' ： 1
    
        
        if linkMap.get(href) == None:
            linkMap[href] = 1
            file.write(href+"\n")
        
    time.sleep(3)
 
    print(list(linkMap.keys()))
file.close()
```

Get links to memes' pages

```python
import pandas as pd
import numpy as np
data = 'https://knowyourmeme.com'
link_arr = []
with open('knowyourmemes.txt', 'r',encoding='utf-8') as f:
    for line in f.readlines():
        line = line.strip()
        test = data + line
        link_arr.append(test)
print(link_arr)
```

Crawl each page for four pieces of information (name,tags,about,img_url)

```python
import requests
import time
from bs4 import BeautifulSoup
import lxml
```

```python
url = 'https://knowyourmeme.com/memes/events/gamergate'

r = urllib.request.urlopen(url)
print(r)
s = BeautifulSoup(r, 'html.parser')

name = s.find('section',class_='info').find('a')['title']
print(name)

tags = s.find('dl',id='entry_tags').find('dd').text
print(tags)

about = s.find('section',class_='bodycopy').find('p').text
print(about)

img_url = s.find('img',title=name)['src']
print(img_url)
```
